```python
import pandas as pd
```

```python
df = pd.read_csv("un-general-debates.csv")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Veri setini ykle (rnek olarak kolonlarn 'year' ve 'speech' olduunu varsayyorum)
# df = pd.read_csv("ungd_data.csv")  # Dosya adn kendi dosyana gre deitir

# Anahtar kelimeler
keywords = [
    "multipolar", "bipolar", "unipolar",
    "emerging powers", "great powers", "superpowers",
    "global south", "new world order", "new order"
]

# Her anahtar kelime iin konuma iinde ka kez getiini say
for keyword in keywords:
    df[keyword] = df["text"].str.lower().str.count(keyword)

# Yllk toplamlar al
yearly_counts = df.groupby("year")[keywords].sum()

# Grselletir
plt.figure(figsize=(14, 8))
for keyword in keywords:
    plt.plot(yearly_counts.index, yearly_counts[keyword], label=keyword, linewidth=1)

plt.title("UN General Debate Speeches: Polarity-Related Term Frequency Over Time")
plt.xlabel("Year")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 8))
for keyword in keywords:
    plt.plot(yearly_counts.index, yearly_counts[keyword], label=keyword, linewidth=10)

plt.title("UN General Debate Speeches: Polarity-Related Term Frequency Over Time")
plt.xlabel("Year")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
